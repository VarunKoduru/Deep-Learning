```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.randn(10,3)
y = torch.randn(10,1)

class SimpleNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3,1)
  def forward(self,x):
    return self.linear(x)

model = SimpleNet()

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(),lr=0.01)

for epoch in range(100):
  y_pred = model(x)
  loss = loss_fn(y_pred,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

print("Final Loss:", loss.item())
```

```python
# AND
def AND(x1,x2):
  w1  = 2
  w2  = 3
  bias = -4
  act = w1*x1 + w2*x2 + bias
  if act >= 0:
    return 1
  else:
    return 0
#OR
def OR(x1,x2):
  w1  = 2
  w2  = 3
  bias = -1
  act = w1*x1 + w2*x2 + bias
  if act >= 0:
    return 1
  else:
    return 0

print(AND(1,0))
print(OR(0,0))
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, regularizers

# Load dataset (top 10,000 words)
vocab_size = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

# Pad sequences to fixed length
max_length = 200
x_train = pad_sequences(x_train, maxlen=max_length)
x_test = pad_sequences(x_test, maxlen=max_length)

# Build ANN model with L2 regularization
def build_l2_model():
    model = models.Sequential([
        layers.Dense(128, activation='relu',
                     input_shape=(max_length,),
                     kernel_regularizer=regularizers.l2(0.001)),

        layers.Dense(64, activation='relu',
                     kernel_regularizer=regularizers.l2(0.001)),

        layers.Dense(1, activation='sigmoid')  # Binary output
    ])
    return model

model_l2 = build_l2_model()

model_l2.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model_l2.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=512,
    validation_split=0.2
)

# Evaluate
test_loss, test_acc = model_l2.evaluate(x_test, y_test)

print(f"Test accuracy: {test_acc:.4f}")
```

```python
# Implement CNN on MNITST Dataset
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras import models, layers

vocab_size = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)


def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

x_train = vectorize_sequences(x_train)
x_test = vectorize_sequences(x_test)


model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(10000,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    epochs=5,
                    batch_size=512,
                    validation_split=0.2)

test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers
import numpy as np

# Set seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

print("TensorFlow version:", tf.__version__)

# ==================== LOAD FASHION MNIST ====================
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalize to [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Add channel dimension: (28, 28, 1)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

print("Dataset shapes:")
print("Train:", x_train.shape, "Test:", x_test.shape)

# ==================== TF.DATA PIPELINES (Memory Efficient) ====================

# 1. For LeNet and ZF-Net (28x28 grayscale)
def preprocess_small(image, label):
    return image, label

train_ds_small = tf.data.Dataset.from_tensor_slices((x_train, y_train)) \
    .map(preprocess_small, num_parallel_calls=tf.data.AUTOTUNE) \
    .shuffle(10000) \
    .batch(128) \
    .prefetch(tf.data.AUTOTUNE)

test_ds_small = tf.data.Dataset.from_tensor_slices((x_test, y_test)) \
    .map(preprocess_small, num_parallel_calls=tf.data.AUTOTUNE) \
    .batch(128) \
    .prefetch(tf.data.AUTOTUNE)

# 2. For Pre-trained models (224x224 RGB - on-the-fly resize)
def preprocess_large(image, label):
    image = tf.image.resize(image, (224, 224))          # Resize
    image = tf.repeat(image, repeats=3, axis=-1)        # Grayscale → 3 channels
    return image, label

# Smaller batch to avoid OOM
train_ds_large = tf.data.Dataset.from_tensor_slices((x_train, y_train)) \
    .map(preprocess_large, num_parallel_calls=tf.data.AUTOTUNE) \
    .shuffle(10000) \
    .batch(32) \
    .prefetch(tf.data.AUTOTUNE)

test_ds_large = tf.data.Dataset.from_tensor_slices((x_test, y_test)) \
    .map(preprocess_large, num_parallel_calls=tf.data.AUTOTUNE) \
    .batch(32) \
    .prefetch(tf.data.AUTOTUNE)

# ==================== MODEL DEFINITIONS ====================

# 1. LeNet-5
def create_lenet():
    model = models.Sequential([
        layers.Conv2D(6, kernel_size=5, activation='relu', input_shape=(28, 28, 1)),
        layers.AveragePooling2D(pool_size=2, strides=2),
        layers.Conv2D(16, kernel_size=5, activation='relu'),
        layers.AveragePooling2D(pool_size=2, strides=2),
        layers.Flatten(),
        layers.Dense(120, activation='relu'),
        layers.Dense(84, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

# 2. ZF-Net (adapted for 28x28)
def create_zfnet():
    model = models.Sequential([
        layers.Conv2D(96, kernel_size=7, strides=2, padding='same', activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D(pool_size=3, strides=2, padding='same'),
        layers.Conv2D(256, kernel_size=5, strides=1, padding='same', activation='relu'),
        layers.MaxPooling2D(pool_size=3, strides=2, padding='same'),
        layers.Conv2D(384, kernel_size=3, padding='same', activation='relu'),
        layers.Conv2D(384, kernel_size=3, padding='same', activation='relu'),
        layers.Conv2D(256, kernel_size=3, padding='same', activation='relu'),
        layers.MaxPooling2D(pool_size=3, strides=2, padding='same'),
        layers.GlobalAveragePooling2D(),
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    return model

# 3. Pre-trained Models (Transfer Learning)
def create_pretrained_model(model_name):
    if model_name == 'vgg':
        base = keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        model_name_str = "VGG16"
    elif model_name == 'inception':
        base = keras.applications.InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        model_name_str = "InceptionV3 (GoogLeNet family)"
    elif model_name == 'resnet':
        base = keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        model_name_str = "ResNet50"
    else:
        raise ValueError("Unsupported model")

    base.trainable = False  # Freeze base for fast training

    model = models.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    return model, model_name_str

# ==================== TRAINING FUNCTION ====================
def train_and_evaluate(model, train_ds, test_ds, epochs=8, model_name="Model"):
    model.compile(
        optimizer=optimizers.Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\n{'='*60}")
    print(f"Training {model_name} for {epochs} epochs...")
    print(f"{'='*60}")

    history = model.fit(
        train_ds,
        epochs=epochs,
        validation_data=test_ds,
        verbose=1
    )

    # Final evaluation
    test_loss, test_acc = model.evaluate(test_ds, verbose=0)
    print(f"\n{model_name} - Final Test Accuracy: {test_acc*100:.2f}%")
    return test_acc * 100

# ==================== RUN ALL MODELS ====================

print("\nStarting training of all models...\n")

# 1. LeNet
lenet = create_lenet()
train_and_evaluate(lenet, train_ds_small, test_ds_small, epochs=10, model_name="LeNet-5")

# 2. ZF-Net
zfnet = create_zfnet()
train_and_evaluate(zfnet, train_ds_small, test_ds_small, epochs=8, model_name="ZF-Net")

# 3. VGG16 (Pre-trained)
vgg_model, vgg_name = create_pretrained_model('vgg')
train_and_evaluate(vgg_model, train_ds_large, test_ds_large, epochs=5, model_name=vgg_name)

# 4. InceptionV3 (GoogLeNet family)
inception_model, inception_name = create_pretrained_model('inception')
train_and_evaluate(inception_model, train_ds_large, test_ds_large, epochs=5, model_name=inception_name)

# 5. ResNet50 (Pre-trained)
resnet_model, resnet_name = create_pretrained_model('resnet')
train_and_evaluate(resnet_model, train_ds_large, test_ds_large, epochs=5, model_name=resnet_name)

print("\n" + "="*60)
print("All models training completed!")
print("="*60)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, utils
from sklearn.metrics import classification_report

# 1. Prepare Data (Subsampling 5000 images to avoid memory crashes)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

def preprocess_zfnet(images, labels, limit=5000):
    images = images[:limit]
    labels = labels[:limit]
    # Resize to 224x224 and repeat to 3 channels (RGB)
    images = tf.image.resize(images[..., tf.newaxis], (224, 224))
    images = tf.repeat(images, 3, axis=-1)
    return images / 255.0, utils.to_categorical(labels, 10)

x_train_zf, y_train_zf = preprocess_zfnet(x_train, y_train)
x_test_zf, y_test_zf = preprocess_zfnet(x_test, y_test, limit=1000)

# 2. Build ZF-Net Architecture
def build_zfnet(input_shape=(224, 224, 3), num_classes=10):
    model = models.Sequential([
        # Layer 1: The "ZF" modification (7x7 kernel, stride 2)
        layers.Conv2D(96, (7, 7), strides=2, padding='same', activation='relu', input_shape=input_shape),
        layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same'),
        layers.Lambda(lambda x: tf.nn.local_response_normalization(x)), # LRN as per paper

        # Layer 2
        layers.Conv2D(256, (5, 5), strides=2, padding='same', activation='relu'),
        layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same'),
        layers.Lambda(lambda x: tf.nn.local_response_normalization(x)),

        # Layers 3, 4, 5
        layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
        layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
        layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D(pool_size=(3, 3), strides=2),

        # Fully Connected Layers
        layers.Flatten(),
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# 3. Train and Evaluate
zf_model = build_zfnet()
zf_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])

print("Starting ZF-Net Training...")
zf_model.fit(x_train_zf, y_train_zf, epochs=5, batch_size=32, verbose=1)

# 4. Performance Metrics
y_pred = zf_model.predict(x_test_zf)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test_zf, axis=1)

print("\n--- ZF-Net Performance Metrics ---")
print(classification_report(y_true, y_pred_classes))
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import models, datasets

# 1. Register the Guided Backprop Gradient
@tf.RegisterGradient("GuidedRelu")
def _GuidedReluGrad(op, grad):
    gate_f = tf.cast(op.outputs[0] > 0, "float32")
    gate_g = tf.cast(grad > 0, "float32")
    return grad * gate_f * gate_g


# 2. Reconstruct Model using the Custom Gradient
from tensorflow.python.framework import ops

def build_guided_model(model):
    # Get the computation graph
    g = tf.compat.v1.get_default_graph()
    with g.gradient_override_map({'Relu': 'GuidedRelu'}):
        # Rebuild the model architecture here (using LeNet as example)
        new_model = models.Sequential([
            layers.Conv2D(16, 3, activation='relu', input_shape=(28, 28, 1)),
            layers.MaxPooling2D(),
            layers.Conv2D(32, 3, activation='relu'),
            layers.Flatten(),
            layers.Dense(10, activation='softmax')
        ])
    return new_model


def compute_guided_backprop(model, image, class_index):
    image = tf.cast(image, tf.float32)
    with tf.GradientTape() as tape:
        tape.watch(image)
        preds = model(image)
        loss = preds[:, class_index]

    # Calculate gradients of the loss w.r.t the input image
    grads = tape.gradient(loss, image)[0]
    return grads

# 3. Execution
(x_train, y_train), _ = datasets.mnist.load_data()
test_img = x_train[0].reshape(1, 28, 28, 1) / 255.0

# Build and get gradients
guided_model = build_guided_model(None) # Usually you'd copy weights from a trained model
saliency = compute_guided_backprop(guided_model, test_img, y_train[0])

# 4. Visualization
saliency = np.abs(saliency) # Take absolute to see magnitude
saliency /= (np.max(saliency) + 1e-5) # Normalize 0 to 1

plt.imshow(saliency.squeeze(), cmap='hot')
plt.title("Guided Backpropagation Saliency Map")
plt.colorbar()
plt.show()
```
