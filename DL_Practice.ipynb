Code to Build a Simple Neural Network with PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.randn(10,3)
y = torch.randn(10,1)

class SimpleNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3,1)
  def forward(self,x):
    return self.linear(x)

model = SimpleNet()

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(),lr=0.01)

for epoch in range(100):
  y_pred = model(x)
  loss = loss_fn(y_pred,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

print("Final Loss:", loss.item())
```
